```python
df.Wine.value_counts(sort=False)
```

```python
df.Wine.value_counts(sort=False).plot.bar(color=['red', 'green', 'blue'])
xlocs, xlabs = plt.xticks()
for i, v in enumerate(df.Wine.value_counts(sort=False)):
    plt.text(xlocs[i] - 0.05, v + 2, str(v))
plt.show()
```
